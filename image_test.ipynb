```python
from nilearn import image

print(image.load_img("Caltech_0051461_func_preproc.nii.gz").shape)
```

```python
first_vol = image.index_img("Caltech_0051461_func_preproc.nii.gz", 42)
print(first_vol.shape)
```

```python
from nilearn import plotting

plotting.plot_stat_map(first_vol, title="First volume")
```

```python
selecte_vols = image.index_img("Caltech_0051461_func_preproc.nii.gz", slice(42, 46))
```

```python
for img in image.iter_img(selecte_vols):
    plotting.plot_stat_map(img)
```
